```python
import pandas as pd
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.stattools import adfuller
```

```python
df=pd.read_csv("Mehra.csv", parse_dates=True)
df
df=df.drop(columns=['date']) # to avoid problmes with diff()
```

```python
def adfullerTest():
    print('adfuller test')
    print('rgnp',adfuller(df["rgnp"].dropna())[1])
    print('pgnp',adfuller(df["pgnp"].dropna())[1])
    print('ulc',adfuller(df["ulc"].dropna())[1])
    print('gdfco',adfuller(df["gdfco"].dropna())[1])
    print('gdf',adfuller(df["gdf"].dropna())[1])
    print('gdfim',adfuller(df["gdfim"].dropna())[1])
    print('gdfcf',adfuller(df["gdfcf"].dropna())[1])
    print('gdfce',adfuller(df["gdfce"].dropna())[1])
```

```python
adfullerTest()
plot_acf(df["rgnp"]);
plot_acf(df["pgnp"]);
plot_acf(df['ulc']);
plot_acf(df['gdfco']);
plot_acf(df['gdf']);
plot_acf(df['gdfim']);
plot_acf(df['gdfcf']);
plot_acf(df['gdfce']);
```

```python
#1 diff

df=df.diff()
adfullerTest()
plot_acf(df["rgnp"].dropna())
plot_acf(df["pgnp"].dropna());
plot_acf(df['ulc'].dropna());
plot_acf(df['gdfco'].dropna());
plot_acf(df['gdf'].dropna());
plot_acf(df['gdfim'].dropna());
plot_acf(df['gdfcf'].dropna());
plot_acf(df['gdfce'].dropna());
```

```python
#2 diff
df=df.diff()
adfullerTest()
plot_acf(df["rgnp"].dropna())
plot_acf(df["pgnp"].dropna());
plot_acf(df['ulc'].dropna());
plot_acf(df['gdfco'].dropna());
plot_acf(df['gdf'].dropna());
plot_acf(df['gdfim'].dropna());
plot_acf(df['gdfcf'].dropna());
plot_acf(df['gdfce'].dropna());
```

Looks like 2nd order differencing is enough for the data to become stationary (max adfuller result is 0.00397)
